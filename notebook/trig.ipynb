```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
```

```python
def dummy_fig():
    fig = plt.figure()
    ax = fig.add_axes([0.1,0.1,0.8,0.8])
    xvals = np.linspace(0,2*np.pi,100)
    cosine = np.cos(xvals)
    sine = np.sin(xvals)
    ax.plot(xvals,np.zeros(100),color=sns.xkcd_rgb['merlot'], linestyle='--')
    ax.plot(xvals,cosine, color=sns.xkcd_rgb['blurple'], label='cosine')
    ax.plot(xvals,sine, color=sns.xkcd_rgb['yellowish orange'], label='sine')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('trig')
    ax.legend()
    return fig,ax

f,a = dummy_fig()
    
    
```
